# iid client_num

```python
log_dir='log/iid/client_num/'
file_ls = scaner_file(log_dir)
for i in range(len(file_ls)):
    print(file_ls[i])
    df=pd.read_csv(file_ls[i],header=None)
    tmp = list(df.iloc[0,:])
    time_idx = tmp.index('time')+1
    shd_idx = tmp.index('SHD')+1
    if 'cleint_num' in tmp:
        group_idx=tmp.index('cleint_num')+1
    elif 'client_num' in tmp:
        group_idx=tmp.index('client_num')+1
    else:
        group_idx=None
    if group_idx!=None:
        for name,group in df.groupby(group_idx): 
            print(f'client_num:{name}\ttime:{group[time_idx].mean():.2f}\tshd:{group[shd_idx].mean():.2f}')
            print()
    else:
        print(f'time:{df[time_idx].mean():.2f}\tshd:{df[shd_idx].mean():.2f}')
        print()
```
